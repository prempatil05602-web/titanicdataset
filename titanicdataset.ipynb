```python
#train test split
#machin learning must be tested on unseen data
#tarin data- model learns
#test data-model is evaluated

#if we test on taining only - model will cheat
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# preprocessing part
# in this part we will apply standardscaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train[:5]
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# x_trained=pd.DataFrame(x_train,columns=x.columns)
# x_trained.head()

x_trained=pd.DataFrame(x_train,columns=x.columns)
x_trained.head()
```

```python
# logistic regration model
# used when output is binary classification

from sklearn.linear_model import LogisticRegression

# creating model

model=LogisticRegression()

# training model

model.fit(x_train,y_train)
# here the ml actual learns patterns
```

```python
#prediction
#predict train and test
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
```

```python
#accurancy check

from sklearn.metrics import accuracy_score
train_acc =accuracy_score(y_train,y_pred_train)
test_acc =accuracy_score(y_test,y_pred_test)


print(train_acc,test_acc)

# accuracy alone is not enough for classification
# if 90 died ,10 survived but the model says everyone died
# the model will be useless

# here is why we are using confusion matrix
```

```python
# proper fit- high train and high test,with a small gap
# overfit-big gap(train>>test)
# underfit-both train and test perform poorly
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_test)
print(cm)

# [tn fp]
# [fn tp]

# tn=true negative-predicted died and actually died
# fp=false positive-predicted survived and actually died
# fn=false negative-predicted died and actually survived
# tp=true positive-predicted survived and actually survived
```

```python
# cofusion matrix is to denote where the model got confused and
# where the model made dangerous mistake
```

```python
# confusion matrix is to denote where the model got confused and
#  where the model made dangerous mistake

# a confusion matrix is a table used only in classification that
#  compares actual and predicted classes, showing the
# number of correct and incorrect predictons, it provieds much
# more insight than accuracy alone and is the foundation for metrics
# like precision , recall and f1-score
```

```python
#visual confusion matrix
sns.heatmap(cm,annot=True,cmap='Blues',fmt='d')
plt.title('confusion matrix')
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
```

```python
#classification report
#precision ,recall,f1 score and support
```

```python
# knn (K Neareser Neighbour)
# KNN is one of the simplest alignment
# it predicts based on nearby data
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
#prediction
y_pred_train_knn=knn.predict(x_train)
y_pred_test_knn=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
train_acc_knn =accuracy_score(y_train,y_pred_train_knn)
test_acc_knn =accuracy_score(y_test,y_pred_test_knn)


print(train_acc_knn,test_acc_knn)
```

```python
print(classification_report(y_test,y_pred_test_knn))
```

```python
for k in range(1,11):
 knn=KNeighborsClassifier(n_neighbors=k)
 knn.fit(x_train,y_train)
 y_pred_test_knn=knn.predict(x_test)
 print(f'accuracy for k={k} is {accuracy_score(y_test,y_pred_test_knn)}')
```
